```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

OFFENSES AGAINST PERSONS

```python
group2 = pd.read_excel(r"C:\Users\Loveline\Desktop\CRIME STATISTICS 2017.xlsx", sheet_name="Offences Against Persons")
group2.set_index("STATES", inplace=True)
group2.drop("TOTAL",axis=0,inplace=True)
group2
```

Murder

```python
group2.groupby("STATES")["MURDER"].max().plot(kind = "barh", figsize = (10,12), color = "red")
plt.minorticks_on()
plt.grid(which="both")
plt.rcParams["font.size"] = 12
plt.title("MURDER CASES IN NIGERIA", fontweight="bold", fontfamily = "algerian",color = "blue")
plt.xlabel("NUMBER OF MURDER CASES", fontfamily = "algerian", color = "b",)
plt.ylabel("STATES", fontfamily = "algerian", color = "b")

plt.savefig('plot.png')

```
